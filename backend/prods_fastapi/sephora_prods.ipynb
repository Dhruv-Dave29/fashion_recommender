```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/the-pudding/data/refs/heads/master/foundation-names/allShades.csv")
df.info()
```

```python
df_filtered = df[~df['url'].str.contains('ulta.com', case=False, na=False)]
df_filtered.info()
```

```python
import math

hex_codes = df_filtered["hex"].tolist()
# print(hex_codes)
monk_skin_dist = {'Monk 10': 0, 'Monk 9': 0, 'Monk 8':0, 'Monk 7':0, 'Monk 6':0, 'Monk 5':0, 'Monk 4':0, 'Monk 3':0, 'Monk 2':0, 'Monk 1':0}


skin_tones = {'Monk 10': '#292420', 'Monk 9': '#3a312a', 'Monk 8':'#604134', 'Monk 7':'#825c43', 'Monk 6':'#a07e56', 'Monk 5':'#d7bd96', 'Monk 4':'#eadaba', 'Monk 3':'#f7ead0', 'Monk 2':'#f3e7db', 'Monk 1':'#f6ede4'}

# Function to convert hex to RGB
def hex_to_rgb(hex_code):
    hex_code = hex_code.lstrip('#')
    return tuple(int(hex_code[i:i+2], 16) for i in (0, 2, 4))

# Function to calculate Euclidean distance between two RGB colors
def euclidean_distance(color1, color2):
    return math.sqrt(sum((c1 - c2) ** 2 for c1, c2 in zip(color1, color2)))

# Iterate through each hex code in the list
for hex_code in hex_codes:
    rgb = hex_to_rgb(hex_code)  # Convert hex to RGB
    min_distance = float('inf')  # Initialize with a large value
    closest_monk = None

    # Compare with each Monk Skin Tone
    for monk, monk_hex in skin_tones.items():
        monk_rgb = hex_to_rgb(monk_hex)
        distance = euclidean_distance(rgb, monk_rgb)

        # Update closest Monk Skin Tone
        if distance < min_distance:
            min_distance = distance
            closest_monk = monk

    # Update the count in monk_skin_dist
    if closest_monk:
        monk_skin_dist[closest_monk] += 1

# Print the result
print(monk_skin_dist)
sum(monk_skin_dist.values())



```

```python

from scipy.spatial import KDTree

def closest_tone_match(rgb_tuple):
    skin_tones = {'Monk 10': '#292420', 'Monk 9': '#3a312a', 'Monk 8':'#604134', 'Monk 7':'#825c43', 'Monk 6':'#a07e56', 'Monk 5':'#d7bd96', 'Monk 4':'#eadaba', 'Monk 3':'#f7ead0', 'Monk 2':'#f3e7db', 'Monk 1':'#f6ede4'}

    rgb_values = []
    names = []
    for monk in skin_tones:
        names.append(monk)
        rgb_values.append(hex_to_rgb(skin_tones[monk]))
    
    kdt_db = KDTree(rgb_values)
    distance, index = kdt_db.query(rgb_tuple)
    # print(rgb_values)
    return names[index]

# match = closest_tone_match((249,230,223))
# print(match)
```
